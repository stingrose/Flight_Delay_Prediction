```
import pandas as pd
import os
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as sns
%matplotlib inline
from collections import Counter 
import matplotlib.pyplot as plt
```

```
dfmerge = pd.read_csv("/content/gdrive/My Drive/finalmerge.csv",index_col=0)
```

```
dfmerge.describe()
```

```
X=dfmerge.drop([
               "ArrDelayMinutes",
                "ArrTime",
                "ArrDel15",
                 "Unnamed: 0_x",
                 "Unnamed: 0_y",
                 "Unnamed: 0.1",
                "Origin",
                "Dest",
                "airport_x",
                "airport_y",
                "FlightDate",
                "date_y",
                "date_x",
                "time_x",
                "time_y",
                "windspeedKmph_y",
                "winddirDegree_y",
                "weatherCode_y",
                "precipMM_y",
                "visibility_y",
                "pressure_y",
                "cloudcover_y",
                "DewPointF_y",
                "WindGustKmph_y",
                "tempF_y",
               "WindChillF_y",
               "humidity_y",],axis=1)
X
```

```
Y=dfmerge["ArrDel15"]
from sklearn.model_selection import train_test_split  
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = None) 
```

```
Y_train.value_counts()
```

# **resampling**

```
X = pd.concat([X_train, Y_train], axis=1)
```

```
not_delayed = X[X.ArrDel15==0]
delayed = X[X.ArrDel15==1]
```

```
from sklearn.utils import resample
delayed_upsampled = resample(delayed,replace=True, n_samples=len(not_delayed),random_state=27)
```

```
upsampled = pd.concat([not_delayed, delayed_upsampled])
```

```
upsampled.ArrDel15.value_counts()
```

```
upsampled
```

```
X = upsampled.drop(['ArrDel15'], axis=1)
y = upsampled.ArrDel15
```

```
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```
X_train
```

```
y_train.value_counts()
```

```
y_test.value_counts()
```

**Decision Tree Classifier**

```
from sklearn.tree import DecisionTreeClassifier

clf=DecisionTreeClassifier()
clf.fit(X_train,y_train)

```

```
y_pred2 = clf.predict(X_test)
```

```
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred2))
```

**Random Forest**

```
from sklearn.ensemble import RandomForestClassifier

# train model
rfc = RandomForestClassifier(n_estimators=10).fit(X_train, y_train)

# predict on test set
rfc_pred = rfc.predict(X_test)
```

```
print(classification_report(y_test, rfc_pred))
```

**EXTRA-TREES Classifier**

```

from sklearn.ensemble import ExtraTreesClassifier
etc=ExtraTreesClassifier(n_jobs=-1,random_state =None)
etc.fit(X_train,y_train)
```

```
Y_predetc=etc.predict(X_test)
```

```
target_names=["Class 0","Class 1"]
from sklearn.metrics import classification_report

print(classification_report(y_test,Y_predetc,target_names=target_names))
```
