```python
import json
import pandas as pd
from pandas.io.json import json_normalize 
```

```python
import os
data = pd.read_csv('weather/weatherlist.txt', sep=",", header=None,delim_whitespace=True)
data=data.transpose()
```

```python
x=data[0].str.split(',', expand=True)
x=x.transpose()
x
```

```python
with open('weather/ATL/2016-1.json') as f:
    d = json.load(f)

#lets put the data into a pandas df
#clicking on raw_nyc_phil.json under "Input Files"
#tells us parent node is 'programs'
nycphil = json_normalize(d['data']['weather'])
#nycphil.to_csv('csc.csv')
nycphil.head(5)
```

```python
# df=[]
dd=pd.DataFrame()
nycphil.to_csv('nyphil.csv')
```

```python
for i in x[0]:
    path='weather/'+i
    for j in os.listdir(path):
        if('2016' in j):
            with open('weather/'+i+'/'+j) as f:
                d = json.load(f)
                data = json_normalize(d['data']['weather'])
                data=data[['date','hourly']]
                #data['airport']=i
                #works_data = json_normalize(data=d['data']['weather'], record_path='hourly', errors='ignore')
                works_data = json_normalize(data=d['data']['weather'], record_path=['hourly'], errors='ignore',meta=['date'])
                works_data['airport'] =i
                dd=dd.append(works_data)
```

```python
finalweatherdata2016=dd[['airport','date','windspeedKmph','winddirDegree','weatherCode','precipMM','visibility',
'pressure',
'cloudcover',
'DewPointF',
'WindGustKmph',
'tempF',
'WindChillF',
'humidity',
'time']]  
```

```python
dd2017=pd.DataFrame()
for i in x[0]:
    path='weather/'+i
    for j in os.listdir(path):
        if('2017' in j):
            with open('weather/'+i+'/'+j) as f:
                d = json.load(f)
                data = json_normalize(d['data']['weather'])
                data=data[['date','hourly']]
                data['airport']=i
                works_data = json_normalize(data=d['data']['weather'], record_path='hourly', errors='ignore',meta=['date'])
                works_data['airport'] =i
                dd2017=dd2017.append(works_data)
```

```python
finalweatherdata2017=dd2017[['airport','date','windspeedKmph','winddirDegree','weatherCode','precipMM','visibility',
'pressure',
'cloudcover',
'DewPointF',
'WindGustKmph',
'tempF',
'WindChillF',
'humidity',
'time']]  
```

```python
finalweatherdata=pd.concat([finalweatherdata2016,finalweatherdata2017])
```
